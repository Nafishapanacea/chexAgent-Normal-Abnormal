```python
orig_data= orig_data[['ImageID', 'ViewPosition_DICOM', 'Projection', 'Labels']]
orig_data= orig_data[orig_data['Projection']=='PA']
orig_data
```

```python
import pandas as pd

txt_path = '/home/jupyter-nafisha/X-ray/CSVs/padchext_0.txt'

filenames = []

with open(txt_path, 'r') as f:
    for line in f:
        parts = line.strip().split()
        if len(parts) > 0 and parts[-1].endswith('.png'):
            filenames.append(parts[-1])

df = pd.DataFrame({'image_id': filenames})
```

```python
selected_files= orig_data.merge(
    df,
    left_on= 'ImageID',
    right_on= 'image_id',
    how= 'inner'
)
```

```python
selected_files= selected_files[['image_id', 'Labels']]
selected_files
```

```python
selected_files.iloc[0:50]
```

```python
# All the samples with label "unchanged" are removed.

import ast

selected_files['Labels'] = selected_files['Labels'].apply(ast.literal_eval)
filtered = selected_files[selected_files['Labels'].apply(lambda x: x == ['unchanged'])]
```

```python
len(filtered)
```

```python
# Remove all rows in 'filtered' from 'selected_files'
selected_files = selected_files[~selected_files.index.isin(filtered.index)]
```

```python
selected_files= selected_files.rename(
    columns={
        'Labels':'class_name'
    }
)
```

```python
selected_files
```

```python
selected_files['label'] = selected_files['class_name'].apply(lambda x: 'Normal' if x == ['normal'] else 'Abnormal')
```

```python
# All the samples with label "exclude" are removed.

filtered_exclude = selected_files[selected_files['class_name'].apply(lambda x: x == ['exclude'])]
```

```python
selected_files = selected_files[~selected_files.index.isin(filtered_exclude.index)]
```

```python
# All the samples with label "chronic changes" are removed.

filtered_chronic = selected_files[selected_files['class_name'].apply(lambda x: x == ['chronic changes'])]
selected_files = selected_files[~selected_files.index.isin(filtered_chronic.index)]
```

```python
selected_files['image_id'][17]
```

```python
# As there is still difficulty while assigning Abnormal label to x ray so the x-rays with the following diseases are considered.
# diseases= VinBig labels + few more (like hernia, pneumonia etc)

allowed_diseases = {
    "aortic enlargement",
    "cardiomegaly",
    "pleural thickening",
    "apical pleural thickening",
    "pneumonia",
    "hiatal hernia",
    "pulmonary fibrosis",
    "nodule/mass",
    "lung opacity",
    "infiltrates",
    "pleural effusion",
    "other lesion",
    "infiltration",
    "ild",
    "calcification",
    "consolidation",
    "atelectasis",
    "pneumothorax",
    "normal"
}

filtered_df = selected_files[
    selected_files["class_name"].apply(
        lambda lst: any(
            disease.lower() in allowed_diseases
            for disease in lst
        )
    )
]
```

```python
filtered_df.to_csv('/home/jupyter-nafisha/X-ray/CSVs/PADCHEST_selected_forInf.csv', index= False)
```
