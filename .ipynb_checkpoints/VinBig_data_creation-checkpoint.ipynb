```python
import os
import pandas as pd

data_folder = "/home/omen/Documents/Nafisha/X-ray-NormalVsAbnormal/train"
csv_path = "Selected_all.csv" 

# Read the CSV
df = pd.read_csv(csv_path)

# Get valid filenames (add .dicom extension)
valid_files = set(df['image_id'].astype(str))

# List all files in the data directory
all_files = os.listdir(data_folder)

deleted = 0

# Loop through and delete unwanted files
for f in all_files:
    file_path = os.path.join(data_folder, f)

    # Delete only if file is not in the CSV list
    if f not in valid_files:
        os.remove(file_path)
        deleted += 1

print(f"Deletion complete! Deleted {deleted} files.")
print(f"Files kept: {len(valid_files)}")
```

```python
import pydicom
from PIL import Image
import numpy as np
import os

count = 0

input_dir = "train"
output_dir = "Data"
os.makedirs(output_dir, exist_ok=True)

for fname in os.listdir(input_dir):
    if fname.lower().endswith(".dicom"):
        dicom_path = os.path.join(input_dir, fname)
        png_path = os.path.join(output_dir, fname.replace(".dicom", ".png"))

        dcm = pydicom.dcmread(dicom_path)

        # JPEG2000 decoding handled automatically if plugins installed
        img = dcm.pixel_array

        img = img.astype(float)
        img = (img - img.min()) / (img.max() - img.min()) * 255.0
        img = img.astype(np.uint8)

        Image.fromarray(img).save(png_path)
        count += 1

    if count % 100 == 0:
        print(count)
    
    # break

print("Conversion completed!")
```
