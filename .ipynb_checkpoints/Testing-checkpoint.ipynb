```python
# vinBig testing data
# test_csv = "/home/jupyter-nafisha/X-ray/CSVs/test.csv"
# img_dir = '/home/common'

# chexpert AP data
# test_csv = "/home/jupyter-nafisha/X-ray/CSVs/chexpert_AP.csv"
# img_dir = '/home/common/chexpert_AP'

# combined test data
test_csv= '/home/jupyter-nafisha/X-ray/CSVs/test_combined.csv'
img_dir = '/home/common'

# NIH for model chexpert+vinBig
# test_csv = '/home/jupyter-nafisha/X-ray/CSVs/NIH_test.csv'
# img_dir = '/home/jupyter-nafisha/X-ray/Inference_data/NIH-test-dataset'

# NIH for model chexpert+vinBig
# test_csv = '/home/jupyter-nafisha/X-ray/CSVs/NLMCXR_test.csv'
# img_dir = '/home/jupyter-nafisha/X-ray/Inference_data/NLMCXR_test'

# padchest dataset
# test_csv = '/home/jupyter-nafisha/X-ray/CSVs/PADCHEST_selected_forInf.csv'
# img_dir = '/home/jupyter-nafisha/X-ray/Inference_data/padchest_selected_dataset'
```

```python
from transforms import get_val_transform
from dataset import XRayDataset
from torch.utils.data import DataLoader

test_dataset = XRayDataset(test_csv, img_dir, transform=get_val_transform())
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False, num_workers=4)

# Ground truth labels from CSV: "Normal" → 0, "Abnormal" → 1
true_labels = test_dataset.df["label"].map({"Normal": 0, "Abnormal": 1}).tolist()
image_names = test_dataset.df["image_id"].tolist()
```

```python
predictions = []
```

```python
with torch.no_grad():
    for images, _ in test_loader:  
        images = images.to(device)

        logits = model(images)             # shape: (B, 1)
        probs = torch.sigmoid(logits).squeeze(1)  # shape: (B,)

        preds = (probs >= 0.5).long()      # thresholding for binary prediction

        predictions.extend(preds.cpu().numpy().tolist())
```

```python
predictions.count(0), predictions.count(1)
```

```python
import pandas as pd

# Save predictions CSV
df = pd.DataFrame({
    "image_name": image_names,
    "true_label": true_labels,
    "predicted_label": predictions
})

df.to_csv("test_predictions.csv", index=False)
# print("Predictions saved to test_predictions.csv")
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

# ---- METRICS ----
acc = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions)
recall = recall_score(true_labels, predictions)      # Recall = Sensitivity
f1 = f1_score(true_labels, predictions)
cm = confusion_matrix(true_labels, predictions)

# Extract TN, FP, FN, TP
tn, fp, fn, tp = cm.ravel()

# Specificity
specificity = tn / (tn + fp)

print("\n==== Evaluation Metrics ====")
print(f"Accuracy               : {acc:.4f}")
print(f"Precision              : {precision:.4f}")
print(f"Recall/Sensitivity     : {recall:.4f}")        # Sensitivity
print(f"Specificity            : {specificity:.4f}")
print(f"F1 Score               : {f1:.4f}")

print("\nConfusion Matrix:")
print(cm)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create confusion matrix
cm = np.array([[tn, fp],
               [fn, tp]])

# Plot
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Predicted Negative", "Predicted Positive"],
            yticklabels=["Actual Negative", "Actual Positive"])
plt.title("Confusion Matrix")
plt.show()
```

```python
import pandas as pd

total_counts = pd.Series({
    'Infiltration': 134,
    'Effusion': 111,
    'Atelectasis': 82,
    'Cardiomegaly': 62,
    'Fibrosis': 59,
    'Nodule': 58,
    'Pneumothorax': 44,
    'Pleural_Thickening': 43,
    'Mass': 42,
    'Emphysema': 35,
    'Consolidation': 23,
    'Hernia': 17
})

fn_counts = pd.Series({
    'Infiltration': 50,
    'Nodule': 20,
    'Atelectasis': 20,
    'Fibrosis': 14,
    'Emphysema': 14,
    'Pleural_Thickening': 12,
    'Cardiomegaly': 9,
    'Effusion': 7,
    'Hernia': 6,
    'Mass': 5,
    'Pneumothorax': 5,
    'Consolidation': 2
})

fn_table = (
    pd.DataFrame({
        'Total': total_counts,
        'FN': fn_counts
    })
    .fillna(0)                          # diseases with 0 FN
    .astype({'FN': int})
)


# Compute FN Rate (%)
fn_table['FN Rate (%)'] = (fn_table['FN'] / fn_table['Total']) * 100
fn_table['FN Rate (%)'] = fn_table['FN Rate (%)'].round(1)

fn_table = fn_table.sort_values('FN Rate (%)', ascending=False)
```
