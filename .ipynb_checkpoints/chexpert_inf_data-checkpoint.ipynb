```python
import pandas as pd
# path= '/home/omen/Documents/Nafisha/X-ray-NormalVsAbnormal/chexpert_AP.csv'
path= '/home/omen/Documents/Nafisha/X-ray-NormalVsAbnormal/chexpert_PA.csv'
data= pd.read_csv(path)
```

```python
data['Path'] = data['Path'].str.replace('CheXpert-v1.0-small/train/','')
```

```python
data
```

```python
import os
import shutil

source_dir = '/home/omen/Documents/Nafisha/X-ray-NormalVsAbnormal/chexpert'
dest_dir   = '/home/omen/Documents/Nafisha/X-ray-NormalVsAbnormal/chexpert_PA'

for _, row in data.iterrows():
    src_path = os.path.join(source_dir, row['Path'])
    dst_path = os.path.join(dest_dir, row['Path'])

    # if not os.path.exists(src_path):
    #     print(src_path)

    os.makedirs(os.path.dirname(dst_path), exist_ok=True)

    shutil.copy2(src_path, dst_path)
    # break
```
